```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Create a meshgrid for the x and y values
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)

# Calculate the Z values
Z = X**2 + Y**2

# Create a 3D plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, cmap='coolwarm')

# Set the x and y limits
ax.set_xlim([-5, 5])
ax.set_ylim([-5, 5])

# Add a title and labels to the axes
ax.set_title('Z = X**2 + Y**2')
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

# Show the plot
plt.show()
```

```python
import pandas as pd
# Matplotlib for additional customization
from matplotlib import pyplot as plt
%matplotlib inline
# Seaborn for plotting and styling
import seaborn as sns
```

```python
!wget -q https://elitedatascience.com/wp-content/uploads/2022/07/Pokemon.csv
# Read dataset
df = pd.read_csv('Pokemon.csv', index_col=0, encoding='latin')

# Display first 5 observations
df.head()
```

```python

# Select the features to calculate the correlation coefficients for
features = ["HP", "Attack", "Defense", "Sp. Atk", "Sp. Def", "Speed"]

# Calculate the Pearson correlation coefficients
pearson_correlations = df[features].corr(method="pearson")

# Calculate the Spearman rank correlation coefficients
spearman_correlations = df[features].corr(method="spearman")

# Plot the heatmaps of the correlation coefficients
plt.subplot(121)
sns.heatmap(pearson_correlations, annot=True)
plt.title("Pearson Correlation")

plt.subplot(122)
sns.heatmap(spearman_correlations, annot=True)
plt.title("Spearman Rank Correlation")

plt.show()
```
