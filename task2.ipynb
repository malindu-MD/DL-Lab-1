```python
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
au_df = pd.read_csv('/content/sample_data/Au_nanoparticle_dataset.csv')
```

```python
print(au_df.shape)
```

```python
for i in range(au_df.shape[0]):
  n_sum = (au_df['N_bulk'][i] + au_df['N_surface'][i])
  assert n_sum == au_df['N_total'][i]
  if(au_df.shape[0]-1) ==i:
    print('all N-bulk and N_surface add up to total')
```

Q3

```python
for feature in ['N_total','N_bulk','N_surface','R_avg']:
  print(f'Feature: {feature}')
  print('Mean:', au_df[feature].mean())
  print('Standard deviation:', au_df[feature].std())
  print('Quartile 1:', au_df[feature].quantile(0.25))
  print('Quartile 3:', au_df[feature].quantile(0.75))
  print('\n')
```
